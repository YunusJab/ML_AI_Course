```python
import pandas as pd

# Step 1: Load the CSV file into a pandas DataFrame
df = pd.read_csv('UK_accident.csv', low_memory=False)

df       
```

```python
print(df.info())
```

```python
# Select and keep only the specified columns in the DataFrame
columns_to_keep = ['Accident_Severity', 'Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions']
df = df[columns_to_keep]
```

```python
print(df.info())
```

```python
# Find unique values in each column
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in {column}: {unique_values}\n")
```

```python
# Dictionary to store the count of unique values for each column
value_counts_dict = {}

# Iterate over each column and count unique values
for column in columns_to_keep:
    value_counts_dict[column] = df[column].value_counts()

# Display the results
for column, counts in value_counts_dict.items():
    print(f"Unique value counts for '{column}':")
    print(counts)
    print("\n" + "-"*50 + "\n")
```

```python
# Reduce the size of the data to 10% of the original data
df_sampled = df.sample(frac=0.1, random_state=42)

# Check the size of the sampled data
print(f"Original data size: {df.shape[0]} rows")
print(f"Reduced data size: {df_sampled.shape[0]} rows")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define the feature columns and the target column
features = ['Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions']
target = 'Accident_Severity'

# One-hot encoding of categorical variables with the updated parameter
encoder = OneHotEncoder(sparse_output=False)
encoded_features = encoder.fit_transform(df[features])

# Convert the target column to a numerical format if needed (it seems to be numerical already)
X = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(features))
y = df[target]  # Labels (Accident_Severity)
```

```python
# Split into training (70%) and test set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Further split training data into train and validation set (80% of 70% = 56% total)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Define the KNN model
knn = KNeighborsClassifier(n_neighbors=5)  

# Train the KNN model
knn.fit(X_train, y_train)

# Predict on the validation set
y_val_pred = knn.predict(X_val)

# Evaluate the model on the validation set
val_accuracy = accuracy_score(y_val, y_val_pred)
print(f"Validation Accuracy: {val_accuracy:.2f}")
```

```python
# Predict on the test set
y_test_pred = knn.predict(X_test)

# Evaluate the model on the Test set
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {test_accuracy:.2f}")
```
